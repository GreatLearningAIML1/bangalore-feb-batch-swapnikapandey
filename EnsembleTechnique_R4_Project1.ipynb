4.Split the dataset into training and test set in the ratio of 70:30 (Training:Test).

```python
# splitting data into training and test set for independent attributes
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn import model_selection
from sklearn import metrics
#from sklearn.decomposition import PCA
from scipy.stats import zscore


X=parkinsons_df.drop(labels="status", axis=1)
y=parkinsons_df['status']
X_std = StandardScaler().fit_transform(X)
train_set, test_set, train_labels, test_labels = train_test_split(X_std, y, test_size=0.30, random_state=2)

```

5. Create the model using entropy method of reducing the entropy and fit it to training data. (5 points)

```python
Dt_model=DecisionTreeClassifier(criterion='entropy',random_state=0)#fully grown tree (high bias error and var error)
```

```python
Dt_model.fit(train_set,train_labels)
```

6. Test the model on test data and what is the accuracy achieved. Capture the predicted values and do a crosstab. (7.5 points)

```python
y_predict_Dt=Dt_model.predict(test_set)
```

```python
y_predict_Dt
```

```python
Dt_acc=metrics.accuracy_score(test_labels,y_predict_Dt)
print(Dt_acc)
```

```python
Dt_cr=metrics.classification_report(test_labels , y_predict_Dt)
print(Dt_cr)
```

7. Use regularization parameters of max_depth, min_sample_leaf to recreate the model. What is the impact on the model accuracy? How does regularization help? (

```python
#Regularized Tree (comparatively var error is reduced)
Dt_model2 = DecisionTreeClassifier(criterion = 'entropy',max_depth=2,min_samples_leaf=2, random_state=0 )
```

```python
Dt_model2.fit(train_set,train_labels)
```

```python
y_predict_Dt2=Dt_model2.predict(test_set)
```

```python
Dt_acc2=metrics.accuracy_score(test_labels,y_predict_Dt2)
print(Dt_acc2)
```

```python
Dt2_cr=metrics.classification_report(test_labels , y_predict_Dt2)
print(Dt2_cr)
```

Here accuracy is increased due to max-depth increase.Bias error is reduced due to feature reduction.

8. Next implement the decision tree using Random Forest. What is the optimal number of trees that gives the best result? (10 points)

```python
#Bootstrapped sampled multiple decision tree(with the hope of controlling bias error)
Rf_model = RandomForestClassifier(n_estimators=11,criterion='entropy',max_depth=4,random_state=0 )
```

```python
Rf_model.fit(train_set, train_labels)
```

```python
y_predict_Rf=Rf_model.predict(test_set)
```

```python
Rf_acc=metrics.accuracy_score(test_labels,y_predict_Rf)
print(Rf_acc)
```

```python
Rf_cr=metrics.classification_report(test_labels , y_predict_Rf)
print(Rf_cr)
```

11 estimators gives best accuracy.

```python
models = []
models.append(('DecisionTree', Dt_model))
models.append(('DT_Regularized',Dt_model2))
models.append(('RandomForest', Rf_model))
```

```python
#k4 cross validation
```

```python
# evaluate each model in turn
results = []
names = []
scoring = 'accuracy'
for name, model in models:
	kfold = model_selection.KFold(n_splits=5,random_state=2)
	cv_results = model_selection.cross_val_score(model, X_std, y, cv=kfold, scoring=scoring)
	results.append(cv_results)
	names.append(name)
	msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
	print(msg)
# boxplot algorithm comparison
fig = plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```
